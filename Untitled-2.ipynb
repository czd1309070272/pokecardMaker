```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, classification_report
plt.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题
# 加载CIFAR-10数据集
(x_train_full, y_train_full), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()

# 将标签从二维压缩成一维
y_train_full = y_train_full.flatten()
y_test = y_test.flatten()

# 查看数据信息
print(f"训练集图片形状: {x_train_full.shape}")  # (50000, 32, 32, 3)
print(f"测试集图片形状: {x_test.shape}")        # (10000, 32, 32, 3)
print(f"训练集标签形状: {y_train_full.shape}")  # (50000,)
print(f"测试集标签形状: {y_test.shape}")        # (10000,)

# CIFAR-10类别名称
class_names = ['飞机', '汽车', '鸟', '猫', '鹿', 
               '狗', '蛙', '马', '船', '卡车']

# 显示几张示例图片
plt.figure(figsize=(10, 5))
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(x_train_full[i])
    plt.title(class_names[y_train_full[i]])
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
# 归一化：将像素值从 0-255 缩放到 0-1
x_train_full = x_train_full.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# 为MLP准备数据：将32x32x3展平为3072维向量
x_train_mlp_full = x_train_full.reshape(x_train_full.shape[0], -1)  # (50000, 3072)
x_test_mlp = x_test.reshape(x_test.shape[0], -1)                    # (10000, 3072)

print(f"MLP输入形状: {x_train_mlp_full.shape}")  # (50000, 3072)

# 划分训练集和验证集（10%作为验证集）
x_train_mlp, x_val_mlp, y_train, y_val = train_test_split(
    x_train_mlp_full, y_train_full,
    test_size=0.1,
    random_state=42,
    stratify=y_train_full
)

# CNN数据保持原始形状 (32, 32, 3)
x_train_cnn_full = x_train_full
x_test_cnn = x_test

x_train_cnn, x_val_cnn, y_train_cnn, y_val_cnn = train_test_split(
    x_train_cnn_full, y_train_full,
    test_size=0.1,
    random_state=42,
    stratify=y_train_full
)

print(f"CNN训练集: {x_train_cnn.shape}")  # (45000, 32, 32, 3)
print(f"CNN验证集: {x_val_cnn.shape}")    # (5000, 32, 32, 3)
```

```python
# 模型1：简单MLP（CIFAR-10版本）
model1_cifar = keras.Sequential([
    layers.Input(shape=(3072,)),           # 输入层：3072个神经元
    layers.Dense(256, activation='relu'),  # 隐藏层
    layers.Dense(10, activation='softmax')  # 输出层：10个类别
])

model1_cifar.summary()

print("=" * 50)
print("模型1 输入输出形状：")
print(f"输入张量形状: (batch_size, 3072)")
print(f"输出张量形状: (batch_size, 10)")
print("=" * 50)

# 编译
model1_cifar.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# 训练
history1_cifar = model1_cifar.fit(
    x_train_mlp, y_train,
    validation_data=(x_val_mlp, y_val),
    epochs=30,
    batch_size=64,
    verbose=1
)

# 评估
test_loss1, test_acc1 = model1_cifar.evaluate(x_test_mlp, y_test, verbose=0)
print(f"\n模型1测试集准确率: {test_acc1:.4f}")
```

```python
from tensorflow.keras import regularizers
from tensorflow.keras.callbacks import ReduceLROnPlateau

# 模型2：改进MLP（加深 + Dropout + 学习率调度）
model2_cifar = keras.Sequential([
    layers.Input(shape=(3072,)),
    
    # 第1隐藏层
    layers.Dense(512, activation='relu'),
    layers.BatchNormalization(),           # 批归一化
    layers.Dropout(0.3),
    
    # 第2隐藏层
    layers.Dense(256, activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.3),
    
    # 第3隐藏层
    layers.Dense(128, activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.3),
    
    # 输出层
    layers.Dense(10, activation='softmax')
])

model2_cifar.summary()

# 学习率调度器
reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=3,
    min_lr=1e-6,
    verbose=1
)

model2_cifar.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history2_cifar = model2_cifar.fit(
    x_train_mlp, y_train,
    validation_data=(x_val_mlp, y_val),
    epochs=40,
    batch_size=64,
    callbacks=[reduce_lr],
    verbose=1
)

test_loss2, test_acc2 = model2_cifar.evaluate(x_test_mlp, y_test, verbose=0)
print(f"\n模型2测试集准确率: {test_acc2:.4f}")
```

```python
from tensorflow.keras import layers, models

# 模型3：CNN for CIFAR-10
model3_cifar = models.Sequential([
    # 第一个卷积块
    layers.Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=(32, 32, 3)),
    layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),
    
    # 第二个卷积块
    layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
    layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),
    
    # 第三个卷积块
    layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
    layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),
    
    # 全连接层
    layers.Flatten(),
    layers.Dense(256, activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.5),
    layers.Dense(10, activation='softmax')
])

model3_cifar.summary()

model3_cifar.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.0005),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history3_cifar = model3_cifar.fit(
    x_train_cnn, y_train_cnn,
    validation_data=(x_val_cnn, y_val_cnn),
    epochs=30,
    batch_size=64,
    verbose=1
)

test_loss3, test_acc3 = model3_cifar.evaluate(x_test_cnn, y_test, verbose=0)
print(f"\n模型3测试集准确率: {test_acc3:.4f}")
```

```python
def plot_learning_curve(history, model_name):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
    
    # 损失曲线
    ax1.plot(history.history['loss'], label='Training Loss', linewidth=2)
    ax1.plot(history.history['val_loss'], label='Validation Loss', linewidth=2)
    ax1.set_title(f'{model_name} - Loss Curve')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Loss')
    ax1.legend()
    ax1.grid(True)
    
    # 准确率曲线
    ax2.plot(history.history['accuracy'], label='Training Accuracy', linewidth=2)
    ax2.plot(history.history['val_accuracy'], label='Validation Accuracy', linewidth=2)
    ax2.set_title(f'{model_name} - Accuracy Curve')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Accuracy')
    ax2.legend()
    ax2.grid(True)
    
    plt.tight_layout()
    plt.show()

# 绘制三个模型的学习曲线
plot_learning_curve(history1_cifar, 'Model 1: Baseline MLP')
plot_learning_curve(history2_cifar, 'Model 2: Improved MLP')
plot_learning_curve(history3_cifar, 'Model 3: CNN')
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题
# 用模型3预测
y_pred = model3_cifar.predict(x_test_cnn)
y_pred_classes = np.argmax(y_pred, axis=1)

# 计算准确率和F1分数
accuracy = accuracy_score(y_test, y_pred_classes)
f1 = f1_score(y_test, y_pred_classes, average='weighted')

print(f"准确率 (Accuracy): {accuracy:.4f}")
print(f"加权F1分数 (Weighted F1-Score): {f1:.4f}")

# 分类报告
print("\n分类报告:")
print(classification_report(y_test, y_pred_classes, target_names=class_names))

# 显示正确/错误分类图片
correct_idx = np.where(y_pred_classes == y_test)[0]
wrong_idx = np.where(y_pred_classes != y_test)[0]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# 正确分类
idx_correct = correct_idx[0]
ax1.imshow(x_test[idx_correct])
ax1.set_title(f'Correct\nTrue: {class_names[y_test[idx_correct]]}\nPred: {class_names[y_pred_classes[idx_correct]]}')
ax1.axis('off')

# 错误分类
idx_wrong = wrong_idx[0]
ax2.imshow(x_test[idx_wrong])
ax2.set_title(f'Incorrect\nTrue: {class_names[y_test[idx_wrong]]}\nPred: {class_names[y_pred_classes[idx_wrong]]}')
ax2.axis('off')

plt.tight_layout()
plt.show()
```

```python
print("=" * 60)
print("CIFAR-10 三个模型结果汇总")
print("=" * 60)
print(f"模型1 (基线MLP):     {test_acc1:.4f}")
print(f"模型2 (改进MLP):     {test_acc2:.4f}")
print(f"模型3 (CNN):         {test_acc3:.4f}")
print("=" * 60)
print(f"CNN vs 基线提升:     {(test_acc3 - test_acc1)*100:.2f}%")
```
